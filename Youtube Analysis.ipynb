##  Channel Statistics

```python
def get_channel_stats(youtube,channel_ids):
    youtube_data = []
    
    request = youtube.channels().list(
                part="snippet,contentDetails,statistics", id=','.join(channel_ids)) #using join() method to make the channel_ids which is a list to join them into a string. 
    
    response = request.execute()
    
   
    
    for i in range(len(response['items'])):
        
        snippet = response['items'][i]['snippet']
        
        statistics = response['items'][i]['statistics']
        
        data = dict(Channel_name = snippet.get('title', 'N/A'),
                    #Description =snippet.get('description','N/A'),
                    ChannelUrl = snippet.get('customUrl','N/A'),
                    PublishedDate = snippet.get('publishedAt','N/A'),
                    Country = snippet.get('country','N/A'),
                    SubscriberCount =  statistics.get('subscriberCount','N/A'), 
                    Views = statistics.get('viewCount','N/A'),
                    VideoCount = statistics.get('videoCount','N/A')
                   )
        
        youtube_data.append(data)
    
    return youtube_data
```

```python
channelInfo = get_channel_stats(youtube,channel_ids)
```

```python
Channel_df = pd.DataFrame(channelInfo)
Channel_df
```
